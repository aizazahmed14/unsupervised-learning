Clustering Algorithms
1. K-Means Clustering:
    K-Means is a popular clustering algorithm.
    It groups data points into clusters based on their similarity.

```python
# importing necessery libraries 

import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


#creating a sample data 

data = {
    'Annual_Income':[15,16,17,28,29,35,45,52,54,68],
    'Spending_Score':[39,81,6,77,40,76,35,65,29,50]
}

df = pd.DataFrame(data)

#visualizing data 
plt.scatter(df['Annual_Income'], df['Spending_Score'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title("Customer Data")
plt.show()

#applyihgn kmean clustering
kmean = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmean.fit_predict(df)

#visualizin the cluster
for cluster in range(3):
    cluster_data = df[df['Cluster']==cluster]
    plt.scatter(cluster_data['Annual_Income'],
                cluster_data['Spending_Score'], label=f'Cluster {cluster}')
    
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('K-means Clustering')
plt.legend()
plt.show()
```

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise) : Unlike K-Means, which requires you to predefine the number of clusters, DBSCAN works by grouping points based on density.

```python
from sklearn.cluster import DBSCAN

#applying DBSCAN
dbscan = DBSCAN(eps=10,min_samples=2)
df['DBSCAN_Cluster'] = dbscan.fit_predict(df[['Annual_Income','Spending_Score']])

#visualizing the DBSCAN Cluster
for cluster in np.unique(df['DBSCAN_Cluster']):
    cluster_data = df[df['DBSCAN_Cluster']==cluster]
    plt.scatter(cluster_data['Annual_Income'],
                cluster_data['Spending_Score'],label=f'Cluster {cluster}')

plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('DBSCAN Clustering')
plt.legend()
plt.show()
```

Association Rule Learning
    
    a technique often used in market basket analysis to find relationships between items

```python
#import library
from mlxtend.frequent_patterns import apriori, association_rules


#creating a dataset
transactions = pd.DataFrame([
    {'Bread':1, 'Butter':1, 'Milk':1},
    {'Bread':1, 'Butter':0, 'Milk':1},
    {'Bread':0, 'Butter':1, 'Milk':1},
    {'Bread':1, 'Butter':1, 'Milk':0}
])

#applyin apriori Algorithm
frequent_itemsets = apriori(transactions, min_support=0.5, use_colnames=True)
print(len(transactions))
#generating association rules
rules = association_rules(frequent_itemsets, metric="confidence",min_threshold=0.7, num_itemsets=4)
print(rules)
```
